```python
import numpy as np
import matplotlib.pyplot as plt
```

## Prerequiste

## Define ODE function: $\displaystyle \frac{dy}{dx} = 2y-x^2-x$

```python
def ODE_func(x,y):
    dydx = 2*y-x**2-x
    return dydx

    
def Slope(x, y, ODE_func):
    Slope = ODE_func(x,y)
    return Slope

h = 0.6
x = np.linspace(1,19,int(18//h+1),dtype=float)
y_initial = 3
```

## 1-1


### Define function Solve_Euler

```python
def Solve_Euler(x, y_initial, ODE_func, h):

    length = int((x[-1] - x[0]) // h + 1)
    x = np.linspace(1,19,int(18//h+1),dtype=float)
    y = np.zeros(length,dtype=float)
    y[0] = y_initial

    for i in range(length-1):
        x[i+1] = x[i] + h
        y[i+1] = y[i] + h*Slope(x[i], y[i], ODE_func)

    return x, y
```

### Using Euler method with h = 0.6

```python
x_solved, y_solved = Solve_Euler(x, y_initial, ODE_func, 0.6)
print(x_solved)
print(y_solved)
```

### Plot the result

```python
plt.plot(x_solved, y_solved)
plt.yscale('log')
plt.grid()
plt.show()
```

## 1-2


### Define function Solve_ModEuler

```python
def Solve_ModEuler(x, y_initial, ODE_func, h):

    length = int((x[-1] - x[0]) // h + 1)
    x = np.linspace(1,19,int(18//h+1),dtype=float)
    y = np.zeros(length,dtype=float)
    y[0] = y_initial

    for i in range(length-1):
        x[i+1] = x[i] + h
        sp1 = Slope(x[i], y[i], ODE_func)
        yEu = y[i] + sp1 * h
        sp2 = Slope(x[i+1], yEu, ODE_func)
        y[i+1] = y[i] + h * (sp1 + sp2) / 2

    return x, y
```

### Using ModEuler with h = 0.6

```python
x_solved2, y_solved2 = Solve_ModEuler(x, y_initial, ODE_func, 0.6)
print(x_solved2)
print(y_solved2)
```

### Plot the result

```python
plt.plot(x_solved2, y_solved2)
plt.yscale('log')
plt.grid()
plt.show()
```

## 1-3

### Define function Solve_RK4

```python
def Solve_RK4(x, y_initial, ODE_func, h):

    length = int((x[-1] - x[0]) // h + 1)
    x = np.linspace(1,19,int(18//h+1),dtype=float)
    y = np.zeros(length,dtype=float)
    y[0] = y_initial

    for i in range(length-1):
        x[i+1] = x[i] + h
        sp1 = Slope(x[i], y[i], ODE_func)
        sp2 = Slope(x[i] + 0.5*h, y[i] + 0.5*sp1*h, ODE_func)
        sp3 = Slope(x[i] + 0.5*h, y[i] + 0.5*sp2*h, ODE_func)
        sp4 = Slope(x[i] + h, y[i] + sp3*h, ODE_func)
        y[i+1] = y[i] + h * (sp1 + 2*sp2 + 2*sp3 + sp4) / 6

    return x, y
```

### Using RK4 with h = 0.6

```python
x_solved3, y_solved3 = Solve_RK4(x, y_initial, ODE_func, 0.6)
print(x_solved3)
print(y_solved3)
```

### Plot the result

```python
plt.plot(x_solved3, y_solved3)
plt.yscale('log')
plt.grid()
plt.show()
```

## 1-4


### Define Analytical solution:  $\displaystyle y = x + 0.5x^2 + e^{2x-2} + 0.5$


```python
def Analytical_solution(x):
    y = x + 0.5*x**2 + np.exp(2*x-2) + 0.5
    return y
```

### Plot true solution and relative errors


```python
y_true = Analytical_solution(x)

plt.plot(x, y_true)
plt.plot(x_solved, y_solved)
plt.plot(x_solved2, y_solved2)
plt.plot(x_solved3, y_solved3)
plt.legend(['True Solution', 'Euler', 'ModEuler', 'RK4'])
plt.yscale('log')
plt.grid()
plt.show()

relative0 = (y_true - y_true)/y_true
relative = (y_true - y_solved)/y_true
relative2 = (y_true - y_solved2)/y_true
relative3 = (y_true - y_solved3)/y_true

plt.plot(x, relative0 * 100)
plt.plot(x, relative * 100)
plt.plot(x, relative2 * 100)
plt.plot(x, relative3 * 100)
plt.ylabel('Relative Error [%]')
plt.legend(['True Solution', 'Euler', 'ModEuler', 'RK4'])
plt.grid()
plt.show()
```
